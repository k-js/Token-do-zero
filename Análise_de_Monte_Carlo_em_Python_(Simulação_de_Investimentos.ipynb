```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros da simulao
np.random.seed(42)  # Reprodutibilidade

initial_investment = 10000  # Investimento inicial em R$
n_years = 10  # Nmero de anos
n_simulations = 1000  # Quantidade de simulaes

# Suposies de retorno mdio e desvio padro anual (ex: aes)
annual_return = 0.08  # 8% ao ano
annual_volatility = 0.15  # 15% de desvio padro

# Simulao
results = []

for i in range(n_simulations):
    prices = [initial_investment]
    for year in range(1, n_years + 1):
        growth = np.random.normal(annual_return, annual_volatility)
        prices.append(prices[-1] * (1 + growth))
    results.append(prices)

results = np.array(results)

# Plotando as simulaes
plt.figure(figsize=(12, 6))
for i in range(100):  # Apenas 100 para no sobrecarregar o grfico
    plt.plot(results[i], color='gray', alpha=0.2)
plt.title('Simulao de Monte Carlo - Crescimento de Investimento')
plt.xlabel('Ano')
plt.ylabel('Valor do Investimento (R$)')
plt.grid(True)
plt.show()

# Estatsticas finais
final_values = results[:, -1]
mean_final = np.mean(final_values)
median_final = np.median(final_values)
percentile_5 = np.percentile(final_values, 5)
percentile_95 = np.percentile(final_values, 95)

print(f'Valor mdio final: R${mean_final:,.2f}')
print(f'Valor mediano final: R${median_final:,.2f}')
print(f'5 percentil: R${percentile_5:,.2f}')
print(f'95 percentil: R${percentile_95:,.2f}')
```
